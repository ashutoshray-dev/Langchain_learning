```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
load_dotenv()
model = ChatOllama(model='qwen2.5:3b')
response = model.invoke("Hello")
response.usage_metadata
```

```python
chunks = []
full_message = None
for chunk in model.stream('Hello'):
    chunks.append(chunk)
    print(chunk.text)
    full_message = chunk if full_message is None else full_message+chunk
print(full_message.content)
```

```python
from langchain.messages import HumanMessage
model = ChatOllama(model='qwen2.5:3b')
human_mess = HumanMessage(content_blocks=[
    {'type':'text', 'text':'Hello'},
    # {'type':'image_url', 'image_url':{'url':'https://media.istockphoto.com/id/177418652/photo/elephants-touching-each-other-gently.jpg?s=612x612&w=0&k=20&c=O4ttRuDzM3F3Ix3nXOKNP4KClFt3fdsbrJ3Xdp9AQ6g='}}
    {'type':'text', 'text':'what is the name of the model currently in use to answer this question?'}
])
response = model.invoke([human_mess])
response
```
