```python
text = """A wormhole is a hypothetical structure that connects disparate points in spacetime. It can be visualized as a tunnel with two ends at separate points in spacetime (i.e., different locations, different points in time, or both). Wormholes are based on a special solution of the Einstein field equations. Wormholes are consistent with the general theory of relativity, but whether they actually exist is unknown. Many physicists postulate that wormholes are merely projections of a fourth spatial dimension, analogous to how a two-dimensional (2D)# being could experience only part of a three-dimensional (3D)# object.

In 1995#, Matt Visser suggested there may be many wormholes in the universe if cosmic strings with negative mass were generated in the early universe. Some physicists, such as Kip Thorne, have suggested how to create wormholes artificially."""
```

```python
# splitter = CharacterTextSplitter(
#     chunk_size = 100,
#     chunk_overlap = 0,
#     separator= '#'
# )
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    separators=['#', ' ', '']
)
result = splitter.split_text(text)
print(len(result))
result
```

```python
text = """import { StateGraph, Annotation, interrupt, Command } from "@langchain/langgraph";
import { MemorySaver } from "@langchain/langgraph";

const GraphState = Annotation.Root({
  input: Annotation<string>,
  approved: Annotation<boolean>
});

const approvalNode = async (state: typeof GraphState.State) => {
  const isApproved = interrupt("Do you approve this request?");
  console.log(`User decision received: ${isApproved}`);
  return { approved: isApproved };
};

const workflow = new StateGraph(GraphState)
  .addNode("agent", (state) => ({ input: "Processed: " + state.input }))
  .addNode("approval", approvalNode)
  .addEdge("__start__", "agent")
  .addEdge("agent", "approval")
  .addEdge("approval", "__end__");

const checkpointer = new MemorySaver();
const app = workflow.compile({ checkpointer });"""
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter, Language
splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.TS,
    chunk_size = 250,
    chunk_overlap = 0
)
result = splitter.split_text(text)
result
```

```python
text = """The salt-cracked cliffs stood as silent sentinels against the relentless push of the Atlantic, their jagged edges softened only by the persistent morning mist. Below, the tide performed its daily retreat, leaving behind a mosaic of sea glass and tangled kelp in the hollows of the tide pools. A single hawk hung suspended on a thermal, its keen eyes scanning the rhythmic pulse of the whitecap waves for any sign of movement amongst the rocks. In this isolated stretch of coastline, the air carried a sharp, bracing chill that seemed to strip away the noise of the modern world, replacing it with the ancient, steady breathing of the sea.

Miles inland, the landscape shifted into a dense tapestry of ancient redwoods and damp earth, where the sunlight struggled to pierce the thick, emerald canopy. Here, the silence was different—heavy with the scent of damp pine needles and the muted scurry of forest life hidden deep within the ferns. Tiny streams cut narrow paths through the moss-covered stones, their gurgling voices providing a constant soundtrack to the towering giants above. While the coast was a place of vast horizons and infinite motion, the forest felt like a cathedral of stillness, holding its breath in the cool, shaded shadows of time."""
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_ollama import OllamaEmbeddings
embedding = OllamaEmbeddings(model='embeddinggemma')
splitter = SemanticChunker(
    embedding, breakpoint_threshold_type="standard_deviation",
    breakpoint_threshold_amount=1
)
result = splitter.create_documents([text])
result
```
